```python
f = open('k8-75.txt', 'r')

s = f.readline()
maxChar = ''
maxCount = 0
Char = '\n'
Count = 0
for c in s:
    if c == Char:
        Count += 1
    else:
        if Count > maxCount:
            maxCount = Count
            maxChar = Char
        Char = c
        Count = 1
        
if Count > maxCount:
    maxCount = Count
    maxChar = Char
print(maxChar, maxCount)
```

```python
cntT = 0
cntF = 0
for i in range(63000000, 75000001):
    divs = set()
    j = i
    while j%2 == 0:
        j //= 2
    if j != int(j**0.5)**2: 
        cntF += 1
        continue
    cntT += 1
    for d in range(1, int(j**0.5) + 1):
        if j%d == 0:
            divs.add(d)
            divs.add(j//d)
    if len(divs) == 5:
        print(i, max(divs))
print(cntT, cntF)
```

```python
def p(n):
    if n == 2:
        return True
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

count = 0
minl = 217438
for i in range(173225, 217438):
    for d in range(2, int(i**0.5)+1):
        if i%d == 0 and p(d) and p(i//d) and d%10 == (i//d)%10:
            count += 1
            if i < minl:
                minl = i

print(count, minl)
```

```python
def f(n, div, count):
    if count >= 3:
        return []
    for i in range(div, int(n**0.5) + 1):
        if n % i == 0:
            n //= i
            if n % i == 0:
                return []
            res = f(n, i + 1, count + 1)
            if len(res) == 0:
                return []
            else:
                return [i] + res
    return [n]

f(165, 2, 0)
count = 0
Sum = 0
for i in range(356712, 420902):
    res = f(i, 2, 0)
    if len(res) == 3 and len(set([k%10 for k in res])) == 1:
        count += 1
        Sum += i
        
print(count, Sum/count)
```
